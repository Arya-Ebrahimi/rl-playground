```python
import gym
from policy_iteration import policy_iteration
from value_iteration import value_iteration
import numpy as np
```

```python
def train(env_name):
    env = gym.make(env_name)

    V_policy, pi_policy = policy_iteration(env=env, gamma=0.9, theta=1e-2)
    V_value, pi_value = value_iteration(env=env, gamma=0.9, theta=1e-2)
    
    return env, V_policy, pi_policy, V_value, pi_value
    
```

```python
def play(env, pi):
    done = False
    state = env.reset()
    env.render()
    total_reward = 0
    while not done:
        action = np.argmax(pi[state])
        next_state, reward, done, info = env.step(action)
        total_reward += reward
        state = next_state
        env.render()
        
    print('total reward is: ', total_reward)
```

### FrozenLake env

tested Policy and Value iteration on FrozenLake Gym env

```python
env, V_policy, pi_policy, V_value, pi_value = train('FrozenLake-v1')
```

```python
play(env, pi_policy)
```

```python
play(env, pi_value)
```

### Taxi env

tested Policy and Value iteration on Taxi Gym env

```python
env, V_policy, pi_policy, V_value, pi_value = train('Taxi-v3')
```
